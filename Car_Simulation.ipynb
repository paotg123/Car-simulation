```python
import agentpy as ap

# To do product 
from itertools import product

#Visualization
import seaborn as sns
import IPython 
import matplotlib.pyplot as plt 
import random
```

```python
POSITIONS_CARS = [(0,26), (0,27), (0,28), # Up
             (49,23), (49,24), (49,25), # Down 
             (26, 0),(27, 0), (28,0), # Left
             (23,49), (24,49), (25,49) # Right
            ]

POSDOWN = [(x,y) for x,y in product(range(0,50), range(26,29))] # DOWN
POSUP = [(x,y) for x,y in product(range(0,50), range(23,26))] # UP
POSLEFT = [(x,y) for x,y in product(range(23,26), range(0,50))] # LEFT
POSRIGHT = [(x,y) for x,y in product(range(26,29), range(0,50))] # RIGHT

DICT_MOVES = {
    
    "TURN-DOWN": {"POSITIONS": POSDOWN, "MOVE":[(1, 0)]},
    "TURN-UP": {"POSITIONS": POSUP, "MOVE":[(-1, 0)]},
    "TURN-LEFT": {"POSITIONS": POSLEFT, "MOVE":[(0,-1)]},
    "TURN-RIGHT": {"POSITIONS": POSRIGHT, "MOVE":[(0, 1)]},
}
```

```python
class CarModel(ap.Model):
    
    def setup(self):
        
        positions_cars = (POSITIONS_CARS)
        #Create agents (cars)
        n_cars = int(self.p['K'])
        cars = self.agents = ap.AgentList(self, n_cars)
        
        #Create agents (streets)
        #positions_streets = [(x,y) for x,y in product(range(50), range(23,29))] + \
        #                    [(x,y) for x,y in product(range(23,29), range(50))]
        
        #streets= self.agent_streets = ap.AgentList(self, len(positions_streets))
        #streets.condition = 0
        
        # Create grid (street)
        self.road = ap.Grid(self, (50, 50), track_empty = True)
        # Add streets 
        #self.road.add_agents(streets, positions= positions_streets, empty= True)
        
        # Add cars
        self.road.add_agents(cars, positions = random.choices(POSITIONS_CARS, k= n_cars), empty = True)
      
        self.agents.condition = 1
     #  self.agent_streets.condition = 0
    
    def step(self):
              
        # Iterate over the agents
        for ag in (self.agents):
            
            # Get current position for the agent
            current_position = self.road.positions[ag]
            
            direction_down = DICT_MOVES['TURN-DOWN']['POSITIONS']
            direction_up = DICT_MOVES['TURN-UP']['POSITIONS']
            direction_left = DICT_MOVES['TURN-LEFT']['POSITIONS']
            direction_right = DICT_MOVES['TURN-RIGHT']['POSITIONS']
            
            
            # moveby indicates the direction of movement
        
    
            if (current_position in (DICT_MOVES['TURN-DOWN']['POSITIONS'])):
                self.road.move_by(ag, DICT_MOVES['TURN-DOWN']['MOVE'][0])
            
            # Move down
            elif (current_position in direction_down):
                moveby = DICT_MOVES['TURN-DOWN']['MOVE'][0]
                self.road.move_by(ag, moveby)
            
            elif (current_position in direction_right):
                moveby =  DICT_MOVES['TURN-RIGHT']['MOVE'][0]
                self.road.move_by(ag, moveby)
                
            elif (current_position in direction_up):
                
                moveby = DICT_MOVES['TURN-UP']['MOVE'][0]
                self.road.move_by(ag, moveby)
            
            elif (current_position in direction_left):
                moveby =  DICT_MOVES['TURN-LEFT']['MOVE'][0]
                self.road.move_by(ag, moveby)
            
    
```

```python
parameters = {
    'K': 6,
    'steps': 100
}
```

```python
def animation_plot(model, ax):
    attr_grid = model.road.attr_grid('condition')
    color_dict = {0: '#B1B1B1', 1: "#ff0000",  None:'#d5e5d5'}
    ap.gridplot(attr_grid, ax=ax, color_dict = color_dict, convert = True)
    ax.set_title(f"Car simulation\n")
    

px = 1/plt.rcParams['figure.dpi']
#fig, ax = plt.subplots()
fig, ax = plt.subplots(figsize=(650*px, 650*px))
model = CarModel(parameters)
animation = ap.animate(model, fig, ax, animation_plot)
IPython.display.HTML(animation.to_jshtml(fps=15))
```
